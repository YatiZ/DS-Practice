```python
import nltk
import pandas as pd 
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
from nltk.tokenize import sent_tokenize
```

```python
from nltk.tokenize import word_tokenize
```

```python
text = """As soon as Winnie reached his spot, he quickly sat down and tried to think of something.\
"Think, think,\" Pooh whispered. So, Pooh thought and thought, but nothing came to mind. As little Pooh 
was busy thinking, Gopher, the beaver suddenly popped out of the ground.
\"If I were you, I would quickly go back home. It\'s Windsday, you know?\""""
```

```python
sentences = sent_tokenize(text)
```

```python
print(len(sentences))
```

```python
words = word_tokenize(text)
```

```python
print(len(words))
```

```python
words_nopunc = []
#empty list to store words without punctuation
```

```python
for w in words:
    if w.isalpha():
        words_nopunc.append(w.lower())
```

```python
print(len(words_nopunc))
```

```python
print(words_nopunc)
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words("english"))
```

```python
filtered_words = []
```

```python
for w in words_nopunc:
    if w not in stopwords:
        filtered_words.append(w)
```

```python
print(len(filtered_words))
```

```python
print(filtered_words)
```

```python
#Ex-5 Frequency Distribution
from nltk.probability import FreqDist
```

```python
#find the frequence
fdist = FreqDist(filtered_words)
```

```python
fdist.most_common(10)
```

```python
import matplotlib.pyplot as plt
```

```python
#plot the frequency distribution graph
fdist.plot(30, cumulative= False)
plt.show
```

```python
#Ex-7 Stemming
from nltk.stem import PorterStemmer
```

```python
stem_words = []
porter = PorterStemmer()
for w in filtered_words:
    stem_words.append(porter.stem(w))
print(stem_words)
```

```python
#Ex-8 Lemmatization
from nltk import WordNetLemmatizer
```

```python
lemma_words = []
lemma = WordNetLemmatizer()
```

```python
import nltk
nltk.download('omw-1.4')
```

```python
for w in filtered_words:
    lemma_words.append(lemma.lemmatize(w,pos="v"))
print(lemma_words)
```

```python
#Ex-9 Part of speech tagging
import nltk
pos_words = nltk.pos_tag(lemma_words)
pos_words
```

```python
#Ex-10 Chunking
grammar = "NP: {<DT>?<JJ>*<NN>}"
```

```python
parser = nltk.RegexpParser(grammar)
```

```python
result = parser.parse(pos_words)
print(result)
```

```python
result.draw()
```

```python
#Ex-12
token_sentences = [word_tokenize(s) for s in sentences]
print(token_sentences)
print("\n")
```

```python
pos_sentences = [nltk.pos_tag(s) for s in token_sentences]
print(pos_sentences)
```

```python
ner_sentences = nltk.ne_chunk_sents(pos_sentences, binary= True)
```

```python
for ner in ner_sentences:
    print(ner)
    print("\n")
    ner.draw()
```
