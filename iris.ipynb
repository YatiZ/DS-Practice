```python
import sklearn
```

```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn import tree
```

```python
from sklearn.tree import export_text
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
x = iris.data 
```

```python
y = iris.target
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(x,y)
```

```python
tree.plot_tree(clf);
```

```python
fn = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
```

```python
cn = ['setosa','versicolor','virginica']
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize = (12,12), dpi = 300)
```

```python
tree.plot_tree(clf, feature_names= fn, class_names=cn, filled= True);
```

```python
fig.savefig('imagename.png')
```

```python
r = export_text(clf, feature_names = fn, show_weights = True)
```

```python
print(r)
```

```python
print(clf.predict([[5.8,2.6,4.0,1.2]]))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state= 100)
```

```python
clf_new = tree.DecisionTreeClassifier()
```

```python
clf_new = clf_new.fit(x_train, y_train)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize = (12,12))
```

```python
tree.plot_tree(clf_new, feature_names= fn, class_names= cn, filled=True);
```

```python
score = clf_new.score(x_test, y_test)
```

```python
print("Accuracy is", score*100)
```

```python
clf_entropy = tree.DecisionTreeClassifier(criterion="entropy", random_state= 100, max_depth=1, min_samples_leaf= 1)
```

```python
clf_entropy = clf_entropy.fit(x_train, y_train)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=1, figsize = (12,12))
```

```python
tree.plot_tree(clf_entropy,feature_names= fn, class_names= cn, filled= True);
```

```python
score_entropy = clf_entropy.score(x_test, y_test)
```

```python
print("Accuracy is", score_entropy*100)
```
